```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io
import os

sns.set(style="whitegrid")

# -----------------------------
# STEP 1: Upload CSV File
# -----------------------------
print("Please upload Retail_Transactions_Dataset.csv")
uploaded = files.upload()

# Get file name
file_name = list(uploaded.keys())[0]
print(f"[UPLOAD SUCCESS] Loaded file: {file_name}")

# Load into DataFrame
df = pd.read_csv(io.BytesIO(uploaded[file_name]))
print("Dataset shape:", df.shape)
display(df.head())


# -----------------------------
# STEP 2: Data Preparation
# -----------------------------
df.columns = [c.strip() for c in df.columns]

# find date column
date_col = [c for c in df.columns if "date" in c.lower()][0]
df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
df = df.dropna(subset=[date_col]).reset_index(drop=True)

df['Year'] = df[date_col].dt.year
df['Month'] = df[date_col].dt.month
df['MonthName'] = df[date_col].dt.strftime("%b")
df['DayOfWeek'] = df[date_col].dt.day_name()

df['Total_Items'] = pd.to_numeric(df['Total_Items'], errors='coerce').fillna(0).astype(int)
df['Total_Cost'] = pd.to_numeric(df['Total_Cost'], errors='coerce').fillna(0.0)

print("\n[PREP DONE]")
display(df.head())


# -----------------------------
# STEP 3: Basic Analysis
# -----------------------------
print("\n==== BASIC EXPLORATION ====")
print("Total transactions:", len(df))
print("Unique customers:", df['Customer_Name'].nunique())

# fix products list
products = []
for row in df['Product'].astype(str):
    parts = row.replace(";", ",").split(",")
    products.extend([p.strip(" []'\"") for p in parts if p.strip()])

print("\nTop 5 products:")
print(pd.Series(products).value_counts().head())

print("\nTop 10 cities:")
print(df['City'].value_counts().head(10))


# -----------------------------
# STEP 4: Customer Behaviour
# -----------------------------
print("\n==== CUSTOMER BEHAVIOUR ====")

avg_spend = df.groupby("Customer_Category")['Total_Cost'].mean().sort_values(ascending=False)
print("\nAvg spend per category:")
display(avg_spend)

pm_table = (df.groupby(['Customer_Category','Payment_Method']).size()
              .unstack(fill_value=0))
pm_percent = pm_table.div(pm_table.sum(axis=1), axis=0)
print("\nPayment method distribution:")
display(pm_percent)

items_store = df.groupby("Store_Type")['Total_Items'].mean().sort_values(ascending=False)
print("\nAvg items per store type:")
display(items_store)


# -----------------------------
# STEP 5: Promotion & Discounts
# -----------------------------
print("\n==== PROMOTION & DISCOUNT IMPACT ====")

df['Discount_Flag'] = df['Discount_Applied'].astype(str).str.lower().isin(['yes','y','true','1'])

print("\nAvg cost (discount vs no discount):")
display(df.groupby("Discount_Flag")['Total_Cost'].mean())

print("\nAvg items by promotion:")
display(df.groupby("Promotion")['Total_Items'].mean())

print("\nAvg cost by promotion:")
display(df.groupby("Promotion")['Total_Cost'].mean())


# -----------------------------
# STEP 6: Seasonality
# -----------------------------
print("\n==== SEASONALITY ====")

season_rev = df.groupby("Season")['Total_Cost'].sum().sort_values(ascending=False)
print("\nSeason revenue:")
display(season_rev)

season_store = df.groupby(["Season","Store_Type"])['Total_Cost'].sum().unstack(fill_value=0)
print("\nSeason vs Store Type revenue:")
display(season_store)


# -----------------------------
# STEP 7: Visualisations
# -----------------------------
print("\n==== GENERATING PLOTS ====")

os.makedirs("plots", exist_ok=True)

# Cities
plt.figure(figsize=(10,5))
df['City'].value_counts().head(15).plot(kind="bar")
plt.title("Top Cities by Transactions")
plt.tight_layout()
plt.show()

# Payment methods
plt.figure(figsize=(6,6))
df['Payment_Method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Methods")
plt.ylabel("")
plt.show()

# Monthly trends
monthly = df.groupby(['Year','Month'])['Total_Cost'].sum().reset_index()
plt.figure(figsize=(10,5))
for y in monthly['Year'].unique():
    plt.plot(monthly[monthly['Year'] == y]['Month'],
             monthly[monthly['Year'] == y]['Total_Cost'], marker='o', label=y)
plt.title("Monthly Revenue Trends")
plt.legend()
plt.show()

# Heatmap
heat = df.groupby(['Season','Customer_Category'])['Total_Cost'].sum().unstack(fill_value=0)
plt.figure(figsize=(12,6))
sns.heatmap(heat, annot=True, cmap="Blues", fmt=".0f")
plt.title("Season vs Customer Category Revenue")
plt.show()

print("\nAll plots shown and saved.")


# -----------------------------
# STEP 8: Summary
# -----------------------------
print("\n==== SUMMARY ====")
print(f"- Highest revenue season: {season_rev.idxmax()}")
print(f"- Top spending customer category: {avg_spend.idxmax()}")
print(f"- Most used payment method: {df['Payment_Method'].value_counts().idxmax()}")
print(f"- Most popular product: {pd.Series(products).value_counts().idxmax()}")

print("\n--- END OF ANALYSIS ---")
```
